# Prepare dataset

```python
benign = pd.read_csv('benign.csv')
malware = pd.read_csv('malware.csv')
```

```python
print(benign.shape, malware.shape)
```

```python
dataset_initial = pd.concat([benign, malware], axis=0)
```

```python
dataset_initial.shape
```

```python
dataset_shuffled = dataset_initial.sample(frac=1).reset_index(drop=True)
dataset_shuffled.shape
```

```python
dataset_initial.head()
```

```python
dataset_shuffled.head()
```

```python
train_df = dataset_shuffled.iloc[:6388, :]
test_df = dataset_shuffled.iloc[6388:, :]
```

```python
print(train_df.shape, test_df.shape)
```

```python
train_df['label'].value_counts()
```

```python
test_df['label'].value_counts()
```

```python
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
